```python
import pandas_datareader as pdr
tsla = pdr.DataReader('tsla',data_source = 'yahoo',start='2020-01-01')
tsla.tail()
```

```python
%matplotlib 
fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))
plt.plot_date(tsla.index,tsla['Close'])
#plt.xticks(rotation = 90)
ax1.grid(True)      #,color = 'g',linewidth=3,linestyle="-")   #we can modify with this
plt.show()
```

```python
def graph(stock,s_date,e_date):
    import pandas_datareader as pdr
    stock_name = pdr.DataReader(stock,data_source = 'yahoo',start=s_date,end=e_date)
    stock_name.tail()
    
    %matplotlib 
    fig = plt.figure()
    ax1 = plt.subplot2grid((1,1),(0,0))
    
    plt.xlabel("dates")
    plt.ylabel("closed")
    
    ax1.plot_date(stock_name.index,stock_name['Close'])
    
    
    ax1.yaxis.label.set_color('g')
    ax1.xaxis.label.set_color("m")
    
    ax1.set_yticks([200,400,600,800,1000])
    
    for labels in ax1.xaxis.get_ticklabels():
        labels.set_rotation(45)
    for a in ax1.yaxis.get_ticklabels():
        a.set_rotation(45)
        
        
    ax1.grid(True)      #,color = 'g',linewidth=3,linestyle="-")   #we can modify with this
    
    
    
    plt.subplots_adjust(top=0.956,
                        bottom=0.205,
                        left=0.114,
                        right=0.971,
                        hspace=0.2,
                        wspace=0.2)
    
    
    plt.show()
    
    return stock_name.head()
```

```python
graph('tsla','2020-01-16','2020-04-16')
```
